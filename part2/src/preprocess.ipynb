```python
# import necessary libraries
import pandas as pd
import numpy as np
```

```python
# import and read files
df1 = pd.read_excel("../sample_data.xlsx", sheet_name="data_1", header=7)
df2 = pd.read_excel("../sample_data.xlsx", sheet_name="data_2", header=7)
df3 = pd.read_excel("../sample_data.xlsx", sheet_name="data_3", header=7)

df1.head()
```

```python
# drop column
df1 = df1.drop(columns=["SN", "DEBIT NOTE"], axis=1)
df2 = df2.drop(columns=["SN", "DEBIT NOTE"], axis=1)
df3 = df3.drop(columns=["SN", "DEBIT NOTE"], axis=1)
```

```python
rename_columns = {
    "INSURED": "insured_name",
    "POLICY NUMBER": "policy_number",
    "PERIOD OF INSURANCE": "insurance_period",
    "SUM INSURED": "sum_insured",
    "PREMIUM": "premium",
    "PPN": "own_retention_ppn",
    "SUM INSURED.1": "own_retention_sum_insured",
    "PREMIUM.1": "own_retention_premium",
    "PPN.1": "treaty_retention_ppn",
    "SUM INSURED.2": "treaty_sum_insured",
    "PREMIUM.2": "treaty_premium",
    "PPN.2": "facultative_outward_ppn",
    "SUM INSURED.3": "facultative_outward_sum_insured",
    "PREMIUM.3": "facultative_outward_premium"
}

df1 = df1.rename(columns=rename_columns)
df2 = df2.rename(columns=rename_columns)
df3 = df3.rename(columns=rename_columns)

# Merge the dataframes
merged_df = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
merged_df
```

```python
merged_df.tail()
```

```python
# split insurance_period into start_date and end_date
merged_df[['insurance_period_start_date', 'insurance_period_end_date']] = merged_df['insurance_period'].str.split(' - ', expand=True)
merged_df.head()
```

```python
# check the data
merged_df.info()
```

```python
# list of columns
merged_df.columns.to_list()
```

```python
# Define the columns and their desired types
desired_columns = {
    "policy_number": "string",
    "insured_name": "string",
    "sum_insured": "float",
    "premium": "float",
    "own_retention_ppn": "float",
    "own_retention_sum_insured": "float",
    "own_retention_premium": "float",
    "treaty_ppn": "float",
    "treaty_sum_insured": "float",
    "treaty_premium": "float",
    "insurance_period_start_date": "datetime64[ns]",
    "insurance_period_end_date": "datetime64[ns]",
}

# Keep only the desired columns (drop others)
merged_df = merged_df[[col for col in desired_columns.keys() if col in merged_df.columns]]

# Convert data types
for col, dtype in desired_columns.items():
    if col in merged_df.columns:
        if dtype == "string":
            merged_df[col] = merged_df[col].astype("string")
        elif dtype == "float":
            merged_df[col] = pd.to_numeric(merged_df[col], errors="coerce")
        elif dtype == "datetime64[ns]":
            merged_df[col] = pd.to_datetime(merged_df[col], errors="coerce")

# Check the result
merged_df.info()
merged_df.head()
```

```python
# shape
merged_df.shape
```

```python
# check for null values
merged_df.isnull().sum()
```

```python
# mode of insurance_period_start_date
mode1 = merged_df["insurance_period_start_date"].mode()
mode2 = merged_df["insurance_period_end_date"].mode()
print(mode1)
print(mode2)
```

```python
# handle rows with missing insurance_period_start_date or insurance_period_end_date
merged_df["insurance_period_start_date"].fillna(mode1.iloc[0], inplace=True)
merged_df["insurance_period_end_date"].fillna(mode2.iloc[0], inplace=True)
merged_df.head()
```

```python
# download the cleaned DataFrame as a CSV file
merged_df.to_csv('processed_insurance.csv', index=False)
```
